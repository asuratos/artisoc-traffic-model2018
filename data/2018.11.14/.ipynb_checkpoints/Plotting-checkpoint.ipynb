Importing packages and initialization of data set arrays

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import os
import glob

sns.set()
outputext = ".png"

try:
    os.mkdir("plots")
    os.mkdir("plots\Traveltime")
    os.mkdir("plots\Asocialness")
    os.mkdir("plots\Carcount")
except:
    pass

asocialvalues = np.linspace(0.0, 1.0, num= 11)
roaddensities = np.arange(10, 210, 10)
reactiontimes = [0.0, 0.5, 1.0] 
```

Importing data sets from csv

```python
datafiles = glob.glob("output*.csv")

temp = []
for file in datafiles:
    temp.append(pd.read_csv(file))

data = pd.concat(temp, ignore_index = True)
    
speeddata = {}
for density in roaddensities:
    speeddata[density] = {}
    
    for asocialness in asocialvalues:
        temp = []
        for rtime in range(3):
            temp.append(pd.read_csv('speeds_%scars_%s_rt%s_ 1000points.csv' % (density, asocialness, rtime)))
        speeddata[density][asocialness] = pd.concat(temp, ignore_index = True)
        


#speedfiles = glob.glob("speeds*.csv")
#for file in speedfiles :
#    temp.append(pd.read_csv(file))
    
#speeddata = pd.concat(temp, ignore_index = True)

#for density in roaddensities:
#    data[density] = {}
    
#    for j, asocialness in enumerate(asocial_values):
#        data[density][asocialness] = pd.read_csv('output_%sasocial%s, 1000points.csv' % (density, asocialness))
        
```

```python
for value in asocialvalues:  
    slowedratio = []
    for density in roaddensities:
        dataslice = data[density]
        for rtime in reactiontimes:
            # count rows for which Slowed = True
            slowed = dataslice[value][dataslice[value].Slowed]["Slowed"].count()
            # only get rows for which Slowed = False
            notslowed = dataslice[value][~dataslice[value].Slowed]["Slowed"].count()
        
            slowedratio.append(slowed / (slowed + notslowed))
    
        sns.lineplot(x = roaddensities, y = slowedratio)#, label = "%s" % (value))
    
plt.show()

Travel time box plots
```
