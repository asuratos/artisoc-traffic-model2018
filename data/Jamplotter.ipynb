```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import os
import glob

sns.set()
outputext = ".png"

try:
    os.mkdir("plots")
    os.mkdir("plots\Jams")
except:
    pass
```

```python
jamdata = pd.read_csv("Jams.csv")
```

```python
jamdata.head()
```

```python
sns.scatterplot(x = "Time", y = "Y", s = 20,
                data = jamdata[(jamdata["Total Cars"] == 250) & (jamdata["Reaction"] == 0) & (jamdata["ASocial Rate"] == 0.5)]
               )

plt.title("Location of Traffic Jams")

plt.xlim(xmin = 0, xmax = 500)
plt.ylim(ymin = 0, ymax = 300)

plt.savefig("fig.png")
plt.show()
```
